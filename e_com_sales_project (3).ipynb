```python
# filing null instead of removing it ...
df["Region"].fillna(df["Region"].mode()[0])
```

```python
df["Sub-Category"].fillna(df["Sub-Category"].mode()[0], inplace= True)
```

```python
df["Product Name"].fillna(df["Product Name"].mode()[0], inplace= True)
```

```python
df["Sales"] = df["Sales"].fillna(df["Sales"].median())
```

```python
df["Sales"]
```

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
```

```python
df["Profit"] = df["Profit"].fillna(df["Profit"].median())
```

```python
df["Discount"] = df["Discount"].fillna(0)
```

```python
df.info()
```

```python
# converstion of date
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)
```

```python
# converstion of date

df["Ship Date"] = pd.to_datetime(
    df["Ship Date"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)
```

```python
# EDA analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/e commerce dataset.csv", encoding='ISO-8859-1')
```

```python
# sales ditribution
sns.histplot(df["Sales"], kde=True)
plt.show()
```

```python
# Category Count
#Which category has highest number of orders.
#Which category has lowest demand.
#Helps understand customer purchasing behavior.
sns.countplot(x=df["Category"])
plt.xticks(rotation=45)
plt.show()
```

```python
#Category-wise Sales and Profit Analysis
category_analysis = df.groupby("Category")[["Sales","Profit"]].sum()
print(category_analysis)
```

```python
category_analysis.plot(kind="bar")
plt.show()
```

```python
#Region-wise Sales and Profit Analysis
region = df.groupby("Region")[["Sales","Profit"]].sum()
print(region)
```

```python
#Impact of Discount on Profitability
sns.scatterplot(x="Discount", y="Profit", data=df)
plt.show()
```

```python
#Positive Relationship Between Sales and Profit
sns.scatterplot(x="Sales", y="Profit", data=df, marker="^")
plt.show()
```

```python
#Which category generates highest average sales. Which category performs poorly.
sns.barplot(x="Category", y="Sales", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month_name()
```

```python
#Monthly Sales Trend
monthly_sales = df.groupby("Order Month")["Sales"].sum()

monthly_sales.plot(kind="line",)
plt.show()
```

```python
#Sales has a moderate positive impact on Profit.
#As discounts increase, profit tends to decrease.
corr = df.select_dtypes(include=np.number).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

Conclusion.
The business shows strong overall sales performance and stable growth.


Profit is not directly proportional to sales in all cases.


Higher discounts significantly reduce profitability.


Some product categories and regions generate higher profits than others.


High sales volume alone does not guarantee high profit.


Product margins and discount strategies have a major impact on profit.


Seasonal and yearly trends indicate stable business operations.


Focusing on high-margin products and controlled discounts can improve overall profitability.


Underperforming categories and regions should be optimized for better business efficiency.
